### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
top_cities = ciudades.nlargest(10, 'population_total')


plt.bar(top_cities['city'], top_cities['population_total'], color='blue')
plt.xticks(rotation=45, ha='center')  # Centrar texto y rotar nombres
plt.title('Ciudades ms pobladas')
plt.xlabel('Ciudad')
plt.ylabel('Poblacin')
plt.show()
```

```python
plt.hist(ciudades['area_total_sq_mi'], bins=50, color='steelblue', alpha=0.7)
plt.title('Distribucin del rea total ocupada', fontsize=18, fontname='Bradley Hand itc')
plt.xlabel('rea (sq mi)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.scatter(ciudades['population_total'], ciudades['area_water_sq_mi'], color='olive', alpha=0.5)
plt.title('Relacin entre poblacin y rea de agua')
plt.xlabel('Poblacin')
plt.ylabel('rea de agua (sq mi)')
plt.show()
```

```python
plt.boxplot(ciudades['elevation_m'].dropna(), vert=False)
plt.title('Distribucin de la altura de las ciudades')
plt.xlabel('Altura (m)')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Grfico de barras
axs[0, 0].bar(top_cities['city'], top_cities['population_total'], color='blue')
axs[0, 0].set_title('Ciudades ms pobladas')
axs[0, 0].tick_params(axis='x', rotation=45)

# Histograma
axs[0, 1].hist(ciudades['area_total_sq_mi'], bins=50, color='steelblue', alpha=0.7)
axs[0, 1].set_title('Distribucin del rea total ocupada')

# Scatter plot
axs[1, 0].scatter(ciudades['population_total'], ciudades['area_water_sq_mi'], color='olive', alpha=0.5)
axs[1, 0].set_title('Poblacin vs rea de agua')

# Boxplot
axs[1, 1].boxplot(ciudades['elevation_m'].dropna(), vert=False)
axs[1, 1].set_title('Distribucin de la altura')

plt.tight_layout()
plt.show()
```
